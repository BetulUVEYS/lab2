# pandas

Pandas is an open-source Python library used for data manipulation and analysis. It provides data structures and functions for quickly and easily working with structured data.

The two main data structures in Pandas are Series and DataFrame. A Series is a one-dimensional array-like object that can hold any data type. A DataFrame is a two-dimensional table-like data structure consisting of rows and columns, similar to a spreadsheet.

Pandas provides a wide range of functionality for handling missing data, merging and joining datasets, grouping data, filtering and selecting data, and handling time-series data, among other things. It also integrates with other popular Python libraries, such as NumPy and Matplotlib.

Overall, Pandas is a powerful tool for data analysis and is widely used in both academia and industry.

```python
#importing the Pandas library.
#about dataframe: print columns, data types for each column, informatin about data and describe it,
import pandas as pd

mydictionary1 = {"name"   :["Ahmet","Ali","Betül","Ayşe","Fatma","Yusuf"],
                 "surname":["Yılmaz","Kaya","Uveys","Demir","Çelik","Şahin"],
                 "midterm":[50,45,65,90,None,30],
                 "final"  :[35,45,70,90,None,70]}

mydataframe1 = pd.DataFrame(mydictionary1) 
print(mydataframe1)
print('\n')
print(mydataframe1.columns)
print('\n')
print(mydataframe1.dtypes)
print('\n')
print(mydataframe1.describe())
print('\n')
print(mydataframe1.info())
```

```python
#loc ,iloc
#loc:print data using given data
#iloc: print data using index value.
print(mydataframe1["name"])
print('\n')
print(mydataframe1.loc[:,"midterm"])#print all of rows in midterm column
print('\n')
print(mydataframe1.loc[:3,"midterm"])#print till row 3 in midterm column
print('\n')
print(mydataframe1.loc[:3, "name":"final"])#print till row 3 in columns between name and final
print('\n')
mydataframe1["yeni_future"] = [1,2,3,4,5,6]
print(mydataframe1.yeni_future)
print('\n')
print(mydataframe1.loc[::])
print('\n')
print(mydataframe1.loc[::-1]) #print the inverse of all rows and all columns
print('\n')
print(mydataframe1.iloc[:,[2]]) #print all rows in index 2 of columns
print('\n')
print(mydataframe1.loc[:,:'midterm'])#print untill row 3 in name column
```

```python
#filtering
filtre1 = mydataframe1.midterm>50                           
mydataframe1["bool"]= filtre1
print(mydataframe1.loc[:,["midterm","bool"]])
print('\n')
filtrelenmis_data= mydataframe1[filtre1] #print filtered data
print(filtrelenmis_data)
print('\n')
filtre2 = mydataframe1.final>50
filtrelenmis_data2 = mydataframe1[filtre2&filtre1] #print the filtered data in two state
print(filtrelenmis_data2)
print('\n')
mydataframe1[mydataframe1.final>50]#print as table 
```

```python
# List Comprehension
import numpy as np
mydictionary2 = {"name"   :["Ahmet","Ali","Betül","Ayşe","Fatma","Yusuf"],
                 "surname":["Yılmaz","Kaya","Uveys","Demir","Çelik","Şahin"],
                 "midterm":[50,45,65,90,None,30],
                 "final"  :[35,45,70,90,None,70]}

mydataframe2 = pd.DataFrame(mydictionary2) 
#delete unknown values
mydataframe2.dropna(inplace=True)  
print(mydataframe2)
print('\n')
#the difference between the mean in numpy and pandas
average1 = mydataframe2.final.mean() #pandas 
print(average1)
print('\n')
average2= np.mean(mydataframe2.final) #numpy 
print(average2)
print('\n')
mydataframe2["average"]= ["below average" if mydataframe2.final.mean()>each else "above average" for each in mydataframe2.final]
print(mydataframe2)
print('\n')
mydataframe2.columns = [each.upper() for each in mydataframe2.columns]
mydataframe2.columns
```
